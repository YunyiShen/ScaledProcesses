```python
lw = 3
bandwith=.1
kernel = 'gaussian'
lo_retain = 100
hi_retain = -100
X_plot = np.linspace(0,1,100).reshape(-1,1)
lns = (0, (4, .5, 1, 1))


# FIGURE 1

lw = 3
bandwith=.1
kernel = 'gaussian'
lo_retain = 100
hi_retain = -100
X_plot = np.linspace(0,1,100).reshape(-1,1)
lns = (0, (4, .5, 1, 1))

cancer_type_names = {}
cancer_type_names['BRCA'] = 'Breast'
cancer_type_names['LGG'] = 'Glioma'
cancer_type_names['LUAD'] = 'Lung'
cancer_type_names['HNSC'] = 'Carcinoma'
cancer_type_names['COADREAD'] = 'Colon'
cancer_type_names['LUSC'] = 'Lung (Sq.)'
cancer_type_names['SKCM'] = 'Melanoma'
cancer_type_names['UCEC'] = 'Uterine'
cancer_type_names['BLCA'] = 'Bladder'
cancer_type_names['OV'] = 'Ovarian'

for N in [5,10,20]:
    plt.figure(figsize = (18,10))
    for c, cancer_type in enumerate(res[N].keys()):
        plt.subplot(2,5,c+1)
        M = res[N][cancer_type]['M']
        ctz = res[N][cancer_type]['cts'][0,-1]
        
        kn = res[N][cancer_type]['cts'][:,N]
        unm = res[N][cancer_type]['cts'][:,-1] - kn

        GD_precision = 1-np.sort(np.abs(res[N][cancer_type]['GD_preds'][:,-1] - unm)/unm)[lo_retain:hi_retain]

        kde = KernelDensity(kernel=kernel, bandwidth=bandwith).fit(GD_precision.reshape(-1,1))
        log_dens = kde.score_samples(X_plot)
        dens = np.exp(log_dens)
        dens = dens/dens.sum()
        plt.plot(X_plot[:, 0], 1-dens.cumsum(), color=color_dictionary['SSB'], lw=lw,
                linestyle='-',  label = 'SSB')
        plt.scatter(X_plot[:, 0][::8], 1-dens.cumsum()[::8], color=color_dictionary['SSB'], lw=lw,
        marker='X', s= 200)


        IBP_precision = 1-np.sort(np.abs(res[N][cancer_type]['IBP_preds'][:,-1] - unm)/unm)[lo_retain:hi_retain]

        kde = KernelDensity(kernel=kernel, bandwidth=bandwith).fit(IBP_precision.reshape(-1,1))
        log_dens = kde.score_samples(X_plot)
        dens = np.exp(log_dens)
        dens = dens/dens.sum()
        plt.plot(X_plot[:, 0], 1-dens.cumsum(), color=color_dictionary['IBP'], lw=lw,
                linestyle='-.', label = '3BB')

        LP_precision = 1-np.sort(np.abs(res[N][cancer_type]['LP_preds'][:,-1] - unm)/unm)[lo_retain:hi_retain]

        kde = KernelDensity(kernel=kernel, bandwidth=bandwith).fit(LP_precision.reshape(-1,1))
        log_dens = kde.score_samples(X_plot)
        dens = np.exp(log_dens)
        dens = dens/dens.sum()
        plt.plot(X_plot[:, 0], 1-dens.cumsum(), color=color_dictionary['LP'], lw=lw,
                linestyle=lns, label = 'LP')

        J_precision = 1-np.sort(np.abs(res[N][cancer_type]['J_preds'][:,-1,-1] - unm)/unm)[lo_retain:hi_retain]

        kde = KernelDensity(kernel=kernel, bandwidth=bandwith).fit(J_precision.reshape(-1,1))
        log_dens = kde.score_samples(X_plot)
        dens = np.exp(log_dens)
        dens = dens/dens.sum()
        plt.plot(X_plot[:, 0], 1-dens.cumsum(), color=color_dictionary['J'][-1], lw=lw,
                linestyle='--', label = 'J4')

        GT_precision = 1-np.sort(np.abs(res[N][cancer_type]['GT_preds'][:,0,0,-1] - unm)/unm)[lo_retain:hi_retain]

        kde = KernelDensity(kernel=kernel, bandwidth=bandwith).fit(GT_precision.reshape(-1,1))
        log_dens = kde.score_samples(X_plot)
        dens = np.exp(log_dens)
        dens = dens/dens.sum()
        plt.plot(X_plot[:, 0], 1-dens.cumsum(), color=color_dictionary['GT'][0], lw=lw,
                linestyle=':', label= 'GT')
        
        

        plt.xlim([.5,1])
        if c in [0,5]:
            plt.ylabel(r'$\mathbb{P}(v_{'+str(N)+'}^{(M)} \geq v)$', fontsize = 20)
        if c>=5:
            plt.xlabel(r'$v$', fontsize = 20)
        plt.title(cancer_type_names[cancer_type]+r', $M=$'+str(M), fontsize = 20)
        plt.xticks(fontsize = 15)
        plt.yticks(fontsize = 15)
    plt.legend(loc='upper center', bbox_to_anchor=(-2, -0.2),fancybox=True, shadow=True, ncol=5, fontsize = 20)
    plt.tight_layout()
    plt.savefig('Final_Plots/small_N_'+str(N)+'_all.pdf', bbox_inches = 'tight', dpi=1000)
    plt.savefig('Final_Plots/small_N_'+str(N)+'_all.png', bbox_inches = 'tight', dpi=1000)
    plt.savefig('Final_Plots/small_N_'+str(N)+'_all.eps', bbox_inches = 'tight', dpi=1000)
    plt.show()
```

```python
for retain_top_percent in [.8]: #[.85]: #[.99, .95, .9]:
    top_percent_s = int(retain_top_percent*1000)
    plt.figure(figsize = (14,6))
    for n, N in enumerate(res):
 
        plt.subplot(3,1,n+1)
        
        IBP_error, GD_error, LP_error = np.zeros([10,1000]), np.zeros([10,1000]), np.zeros([10,1000])
        J_error, GT_error = np.zeros([10,4,1000]), np.zeros([10,2,1000])
        
        for c, cancer in enumerate(res[N]):
        
            kn = res[N][cancer]['cts'][:,N]
            unm = res[N][cancer]['cts'][:,-1] - kn

            IBP_error[c] =  1-np.sort(np.clip(np.abs(res[N][cancer]['IBP_preds'][:,-1]-unm)/unm , 0, 1))
            GD_error[c] =  1-np.sort(np.clip(np.abs(res[N][cancer]['GD_preds'][:,-1]-unm)/unm , 0, 1))
            LP_error[c] =  1-np.sort(np.clip(np.abs(res[N][cancer]['LP_preds'][:,-1]-unm)/unm , 0, 1))
            
            for order in range(4):
                J_error[c,order] = 1-np.sort(np.clip(np.abs(res[N][cancer]['J_preds'][:,order,-1]-unm)/unm , 0, 1))
            for gt_order in range(2):
                GT_error[c, gt_order] = 1-np.sort(np.clip(np.abs(res[N][cancer]['GT_preds'][:,gt_order,0,-1]-unm)/unm , 0, 1))


        IBP = np.ndarray.flatten(IBP_error[:, :top_percent_s])
        GD = np.ndarray.flatten(GD_error[:, :top_percent_s])
        LP = np.ndarray.flatten(LP_error[:, :top_percent_s])
        J1 = np.ndarray.flatten(J_error[:, 0, :top_percent_s])
        J2 = np.ndarray.flatten(J_error[:, 1, :top_percent_s])
        J3 = np.ndarray.flatten(J_error[:, 2, :top_percent_s])
        J4 = np.ndarray.flatten(J_error[:, 3, :top_percent_s])
        GT1 = np.ndarray.flatten(GT_error[:, 0, :top_percent_s])
        GT2 = np.ndarray.flatten(GT_error[:, 1, :top_percent_s])
        
        labs = ['$SSB$ ('+str(np.median(GD)*100)[:4]+'%)', r'$3$BB ('+str(np.median(IBP)*100)[:4]+'%)', 'LP ('+str(np.median(LP)*100)[:4]+'%)', r'$J_1$ ('+str(np.median(J1)*100)[:4]+'%)', r'$J_2$ ('+str(np.median(J2)*100)[:4]+'%)', r'$J_3$ ('+str(np.median(J3)*100)[:4]+'%)', r'$J_4$ ('+str(np.median(J4)*100)[:4]+'%)', r'$GT_1$ ('+str(np.median(GT1)*100)[:4]+'%)', r'$GT_2$ ('+str(np.median(GT2)*100)[:4]+'%)']
        plt.ylabel('Accuracy', fontsize = 20)
        bp = plt.boxplot([GD, IBP, LP, J1, J2, J3, J4, GT1, GT2], labels=labs, **black_dict)
        for patch, color in zip(bp['boxes'], colors): patch.set_facecolor(color) 
        for cap in bp['medians']: cap.set(linewidth = 2) 
        for whisker in bp['whiskers']: whisker.set(linewidth = 1.5)
        for cap in bp['caps']: cap.set(linewidth = 2)
        plt.yticks(np.linspace(0,1,5), [str(a)+'%' for a in np.linspace(1,100,5,dtype = int)])
        plt.xticks(fontsize = 15)
        plt.yticks(fontsize = 15)
        plt.title(r'$N = $'+str(int(N)), fontsize = 20)
        plt.ylim([0,1.1])
        # plt.suptitle('IBP', y = 1.04, fontsize = 30)
    plt.tight_layout()
    plt.savefig('Final_Plots/small_N_paper_all_cancer_boxplot.pdf', dpi = 1000, bbox_inches = 'tight')
    plt.savefig('Final_Plots/small_N_paper_all_cancer_boxplot.eps', dpi = 1000, bbox_inches = 'tight')
    plt.show()
```

```python
threshold_ls = [1, .9, .75, .2, .1]

for threshold in threshold_ls:
    plt.figure(figsize = (20,4))
    for f, fraction in enumerate(results.keys()):
        plt.subplot(1,len(results.keys()), f+1)
        GD_cov, IBP_cov = np.zeros([error_dictionary['percentage_checks'].shape[0], len(results[.05].keys())]), np.zeros([error_dictionary['percentage_checks'].shape[0], len(results[.05].keys())])
        for c, cancer in enumerate(results[fraction]):

            if min(error_dictionary_prediction[fraction][cancer]['GD_error'])<threshold:
                GD_indices = np.where(error_dictionary_prediction[fraction][cancer]['GD_error']<threshold)[0]
                GD_cov[:,c] = error_dictionary[fraction][cancer]['GD_all'][:, GD_indices].mean(axis = 1)
            
            if min(error_dictionary_prediction[fraction][cancer]['IBP_error'])<threshold:
                IBP_indices = np.where(error_dictionary_prediction[fraction][cancer]['IBP_error']<threshold)[0]
                IBP_cov[:,c] = error_dictionary[fraction][cancer]['IBP_all'][:, IBP_indices].mean(axis = 1)
        plt.title(r'N/(N+M) = '+str(int(fraction*100))+'%', fontsize = 20)
        plt.scatter(error_dictionary['percentage_checks'],error_dictionary['percentage_checks'], color = 'k')
        
        for _ in range(GD_cov.shape[1]):
            plt.scatter(error_dictionary['percentage_checks']-.02, GD_cov[:,_], color = color_dictionary['SSB'], marker = 'X', alpha  =.05)
            plt.scatter(error_dictionary['percentage_checks']+.02, IBP_cov[:,_], color = color_dictionary['IBP'], marker = 'v', alpha  =.05)
        plt.plot(error_dictionary['percentage_checks'], GD_cov.mean(axis = -1), color = color_dictionary['SSB'], label = 'SSB', linestyle ='-.', linewidth=4)
        plt.plot(error_dictionary['percentage_checks'], IBP_cov.mean(axis = -1), color = color_dictionary['IBP'], label = '3BB', linestyle =':', linewidth=4)
        
        plt.plot(error_dictionary['percentage_checks'], error_dictionary['percentage_checks'], linewidth = 2, color = 'k')
        plt.yticks(error_dictionary['percentage_checks'][::3], np.array(100*error_dictionary['percentage_checks'][::3],dtype = int), fontsize = 20)
        plt.xticks(error_dictionary['percentage_checks'][::3], np.array(100*error_dictionary['percentage_checks'][::3],dtype = int), fontsize = 20)

        plt.title(r'N/(N+M) = '+str(int(fraction*100))+'%', fontsize = 26)
        plt.ylim([-.05,1.1])
        if f == 0:
            plt.ylabel('Coverage', fontsize = 24)
            plt.legend(fontsize = 22)
        plt.xlabel('Width of credible interval', fontsize = 22)
    plt.tight_layout()
    plt.savefig('Final_Plots/coverage_all_'+str(threshold)+'.pdf', dpi = 1000, bbox_inches = 'tight')
    plt.savefig('Final_Plots/coverage_all_'+str(threshold)+'.eps', dpi = 1000, bbox_inches = 'tight')
    plt.show()
```
