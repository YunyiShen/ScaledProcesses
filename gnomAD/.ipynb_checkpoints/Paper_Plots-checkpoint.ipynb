```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
from tqdm import tqdm_notebook
import matplotlib.pyplot as plt
#import matplotlib
import pyreadr
import sys

sys.path.append('../utils_folder/')
from scipy.stats import norm
from sklearn.neighbors import KernelDensity
from matplotlib import rc as mtprc
mtprc('xtick', labelsize=20) 
mtprc('ytick', labelsize=20)

from utils_GD import *
from utils_IBP import *
from utils_plots import *

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
orange_dict = make_color_dict('orange')
black_dict = make_color_dict('k')
```

```python
res = np.load('results/_new_all_populations_regression.npy', allow_pickle=1).item()
color_dictionary = np.load('Plots/color_dictionary.npy', allow_pickle=1).item()
colors =[ color_dictionary['SSB'], color_dictionary['IBP'], color_dictionary['LP'], color_dictionary['J'][0], color_dictionary['J'][1], color_dictionary['J'][2], color_dictionary['J'][3], color_dictionary['GT'][0], color_dictionary['GT'][1]]
```

```python

lw = 3
bandwith=.1
kernel = 'gaussian'
lo_retain = 0
hi_retain = -1
X_plot = np.linspace(0,1,100).reshape(-1,1)
lns = (0, (4, .5, 1, 1))


# FIGURE 1

lw = 3
bandwith=.1
kernel = 'gaussian'
lo_retain = 100
hi_retain = -100
X_plot = np.linspace(0,1,100).reshape(-1,1)
lns = (0, (4, .5, 1, 1))

population_names = {}
population_names['amr'] = 'Amr.'
population_names['eas'] = 'SE. As.'
population_names['eas_oea'] = 'Ot. E. As.'
population_names['fin'] = 'Fin.'
population_names['nfe_seu'] = 'S. Eu.'
population_names['nfe_swe'] = 'Swe.'
population_names['sas'] = 'S. As.'
population_names['oth'] = 'Other'
```

```python
for N in [50, 100]: 
    plt.figure(figsize = (18,12))
    for p, population in enumerate(population_names):
        plt.subplot(2,4,p+1)

        
        N_M = -1
        kn = res[N][population]['K']
        unm = res[N][population]['K_new'] 

        GD_precision = res[N][population]['GD_precision']

        kde = KernelDensity(kernel=kernel, bandwidth=bandwith).fit(GD_precision.reshape(-1,1))
        log_dens = kde.score_samples(X_plot)
        dens = np.exp(log_dens)
        dens = dens/dens.sum()
        plt.plot(X_plot[:, 0], 1-dens.cumsum(), color=color_dictionary['SSB'], lw=lw,
                linestyle='-',  label = 'SSB')
        plt.scatter(X_plot[:, 0][::8], 1-dens.cumsum()[::8], color=color_dictionary['SSB'], lw=lw,
        marker='X', s= 200)


        IBP_precision = res[N][population]['IBP_precision'] #[lo_retain:hi_retain]

        kde = KernelDensity(kernel=kernel, bandwidth=bandwith).fit(IBP_precision.reshape(-1,1))
        log_dens = kde.score_samples(X_plot)
        dens = np.exp(log_dens)
        dens = dens/dens.sum()
        plt.plot(X_plot[:, 0], 1-dens.cumsum(), color=color_dictionary['IBP'], lw=lw,
                linestyle='-.', label = '3BB')

        LP_precision = res[N][population]['LP_precision'] #[lo_retain:hi_retain]

        kde = KernelDensity(kernel=kernel, bandwidth=bandwith).fit(LP_precision.reshape(-1,1))
        log_dens = kde.score_samples(X_plot)
        dens = np.exp(log_dens)
        dens = dens/dens.sum()
        plt.plot(X_plot[:, 0], 1-dens.cumsum(), color=color_dictionary['LP'], lw=lw,
                linestyle=lns, label = 'LP')

        J_precision = res[N][population]['J_precision'][:,-1] #[lo_retain:hi_retain]

        kde = KernelDensity(kernel=kernel, bandwidth=bandwith).fit(J_precision.reshape(-1,1))
        log_dens = kde.score_samples(X_plot)
        dens = np.exp(log_dens)
        dens = dens/dens.sum()
        plt.plot(X_plot[:, 0], 1-dens.cumsum(), color=color_dictionary['J'][-1], lw=lw,
                linestyle='--', label = 'J4')

        GT_precision = res[N][population]['GT_precision'][:,0] #[lo_retain:hi_retain]

        kde = KernelDensity(kernel=kernel, bandwidth=bandwith).fit(GT_precision.reshape(-1,1))
        log_dens = kde.score_samples(X_plot)
        dens = np.exp(log_dens)
        dens = dens/dens.sum()
        plt.plot(X_plot[:, 0], 1-dens.cumsum(), color=color_dictionary['GT'][0], lw=lw,
                linestyle=':', label= 'GT')



        plt.xlim([.5,1])
        if p in [0,4]:
            plt.ylabel(r'$\mathbb{P}(v_{'+str(N)+'}^{(M)} \geq v)$', fontsize = 20)
        if p>=5:
            plt.xlabel(r'$v$', fontsize = 20)
        M = res[N][population]['M']
        plt.title(population_names[population]+r', $M=$'+str(M), fontsize = 20)
        plt.xticks(fontsize = 15)
        plt.yticks(fontsize = 15)
    plt.legend(loc='upper center', bbox_to_anchor=(-1.5, -0.2),fancybox=True, shadow=True, ncol=5, fontsize = 20)
    plt.tight_layout()
#     plt.savefig('Plots/gnomAD_small_N_'+str(N)+'_M_'+str(M)+'_all.pdf', bbox_inches = 'tight', dpi=1000)
#     plt.savefig('Plots/gnomAD_small_N_'+str(N)+'_M_'+str(M)+'_all.png', bbox_inches = 'tight', dpi=1000)
#     plt.savefig('Plots/gnomAD_small_N_'+str(N)+'_M_'+str(M)+'_all.eps', bbox_inches = 'tight', dpi=1000)
    plt.show()
```

```python
# N = 100
M = -1
top_percent_s = -1
for N in [50,100]:
    

    IBP_error, GD_error, LP_error = np.zeros([len(population_names),50]), np.zeros([len(population_names),50]), np.zeros([len(population_names),50])
    J_error, GT_error = np.zeros([len(population_names),4,50]), np.zeros([len(population_names),2,50])

    for p, population in enumerate(population_names):

        IBP_error[p] = res[N][population]['IBP_precision']
        GD_error[p] =  res[N][population]['GD_precision']
        LP_error[p] =  res[N][population]['LP_precision']

        for order in range(4):
            J_error[p,order] = res[N][population]['J_precision'][:,order]
        for gt_order in range(2):
            GT_error[p, gt_order] = res[N][population]['GT_precision'][:,gt_order]

    plt.figure(figsize = (20,4))
    IBP = np.ndarray.flatten(IBP_error[:, :top_percent_s])
    GD = np.ndarray.flatten(GD_error[:, :top_percent_s])
    LP = np.ndarray.flatten(LP_error[:, :top_percent_s])
    J1 = np.ndarray.flatten(J_error[:, 0, :top_percent_s])
    J2 = np.ndarray.flatten(J_error[:, 1, :top_percent_s])
    J3 = np.ndarray.flatten(J_error[:, 2, :top_percent_s])
    J4 = np.ndarray.flatten(J_error[:, 3, :top_percent_s])
    GT1 = np.ndarray.flatten(GT_error[:, 0, :top_percent_s])
    GT2 = np.ndarray.flatten(GT_error[:, 1, :top_percent_s])

    labs = ['$SSB$ ('+str(np.median(GD)*100)[:4]+'%)', r'$3$BB ('+str(np.median(IBP)*100)[:4]+'%)', 'LP ('+str(np.median(LP)*100)[:4]+'%)', r'$J_1$ ('+str(np.median(J1)*100)[:4]+'%)', r'$J_2$ ('+str(np.median(J2)*100)[:4]+'%)', r'$J_3$ ('+str(np.median(J3)*100)[:4]+'%)', r'$J_4$ ('+str(np.median(J4)*100)[:4]+'%)', r'$GT_1$ ('+str(np.median(GT1)*100)[:4]+'%)', r'$GT_2$ ('+str(np.median(GT2)*100)[:4]+'%)']
    plt.ylabel('Accuracy', fontsize = 20)
    bp = plt.boxplot([GD, IBP, LP, J1, J2, J3, J4, GT1, GT2], labels=labs, **black_dict)
    for patch, color in zip(bp['boxes'], colors): patch.set_facecolor(color) 
    for cap in bp['medians']: cap.set(linewidth = 2) 
    for whisker in bp['whiskers']: whisker.set(linewidth = 1.5)
    for cap in bp['caps']: cap.set(linewidth = 2)
    plt.yticks(np.linspace(0,1,5), [str(a)+'%' for a in np.linspace(1,100,5,dtype = int)])
    plt.xticks(fontsize = 15)
    plt.yticks(fontsize = 15)
    plt.title(r'$N = $'+str(int(N)), fontsize = 20)
    plt.ylim([0,1.1])
    # plt.suptitle('IBP', y = 1.04, fontsize = 30)
    plt.tight_layout()
#     plt.savefig('Plots/N_'+str(N)+'_paper_all_cancer_boxplot.pdf', dpi = 1000, bbox_inches = 'tight')
#     plt.savefig('Plots/N_'+str(N)+'_paper_all_cancer_boxplot.eps', dpi = 1000, bbox_inches = 'tight')
    plt.show()
```
